```python
import os # os = operating system
```

```python
from PIL import Image # 대문자로 시작하는 Image 모듈 import

root_dir = './images/images'

files = os.path.join(root_dir)
File_names = os.listdir(files)

print('File_names', File_names)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(2,3, figsize=(15,8))
first_five = File_names[:6]

def subplots(): # i, j, k 초기값 0으로 지정
  i = 0
  j = 0
  k = 0
  for k in range(5):
    state = os.path.join(root_dir,first_five[k])
    img = Image.open(state)
    axes[i,j].imshow(img)

    if k==2:
      i += 1
      j = 0
    else:
      j += 1

    plt.tight_layout(pad=2)

subplots()
```
